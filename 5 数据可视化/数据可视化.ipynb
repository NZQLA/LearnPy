```python
import matplotlib.pyplot as plt
import numpy as np

# 
x = np.linspace(0, 10, 100)
# y = np.sin(x)
# y = np.cos(x)

# # 
# plt.figure()

# # 
# plt.plot(x, y)

# # 
# # plt.title('Sine Wave')
# plt.title('Cosine Wave')
# plt.xlabel('X Axis')
# plt.ylabel('Y Axis')

# 
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
y_cot = 1 / np.tan(x)
#  tan  cot 
y_tan = np.clip(y_tan, -10, 10)
y_cot = np.clip(y_cot, -10, 10)

# 
plt.figure()

# 
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.plot(x, y_tan, label='tan(x)')
plt.plot(x, y_cot, label='cot(x)')

# 
plt.title('Trigonometric Functions')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#  y 
plt.ylim(-2, 2)

# 
plt.legend()

# 
plt.show()
```
