```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一些示例数据
x = np.linspace(0, 10, 100)
# y = np.sin(x)
# y = np.cos(x)

# # 创建一个图形
# plt.figure()

# # 绘制数据
# plt.plot(x, y)

# # 添加标题和标签
# # plt.title('Sine Wave')
# plt.title('Cosine Wave')
# plt.xlabel('X Axis')
# plt.ylabel('Y Axis')

# 计算不同的函数值
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
y_cot = 1 / np.tan(x)
# 限制 tan 和 cot 的值
y_tan = np.clip(y_tan, -10, 10)
y_cot = np.clip(y_cot, -10, 10)

# 创建一个图形
plt.figure()

# 绘制不同的函数曲线
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.plot(x, y_tan, label='tan(x)')
plt.plot(x, y_cot, label='cot(x)')

# 添加标题和标签
plt.title('Trigonometric Functions')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# 设置 y 轴范围
plt.ylim(-2, 2)

# 添加图例
plt.legend()

# 显示图像
plt.show()
```
